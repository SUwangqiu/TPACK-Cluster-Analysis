```python
import pandas as pd
```

```python
df = pd.read_spss("CompleteTPACK_Pre_361_no1026_1009ForMplus.sav")
```

```python
df_pre=df.iloc[:,2:34]
```

```python
df_pre
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(df_pre)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
sns.set()
```

```python
plt.plot(range(1, 11), wcss)
plt.title('Selecting the Numbeer of Clusters using the Elbow Method')
plt.xlabel('Clusters')
plt.ylabel('WCSS')
```

```python
kmeans = KMeans(n_clusters=5, random_state=0)
five_cluster=kmeans.fit(df_pre)
```

```python
five_cluster.cluster_centers_
```

```python
column_names = list(df_pre.columns)
```

```python
centers=pd.DataFrame(five_cluster.cluster_centers_, columns=column_names)
```

```python
centers.to_csv("fiveCenters.csv", index =False)
```
